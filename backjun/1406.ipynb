```python
string = input()
left, right = [], []
for c in string:
    left.append(c)
n = int(input())
for _ in range(n):
    c = input()
    if c == "P":
        text = input()
        left.append(text)
    elif c == "L":
        text = left.pop()
        right.insert(0,text)
    elif c == "D":
        text = right.pop
        left.append(text)
    elif c == "B":
        left.pop() 
    
for l in left:
    print(l, end="")
    
for r in right:
    print(r, end="")
    
    
        
```

```python
string_list = list(input())
cursor = len(string_list)

for _ in range(int(input())):
    command = list(input().split())
    if command[0] == 'P':
        string_list.insert(cursor, command[1])
        cursor += 1
  
    elif command[0] == 'L':
        if cursor > 0:
            cursor -= 1

    elif command[0] =='D':
        if cursor < len(string_list):
            cursor += 1
  
    else:
        if cursor > 0:
            string_list.remove(string_list[cursor-1])

print(''.join(string_list))
```
