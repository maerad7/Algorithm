- Method
1. left, right 리스트 생성
2. 커서 왼쪽은 left, 커서 오른쪽은 right
3. 맨 처음 입력 하고 나면 커서는 오른쪽에 있으므로 left 리스트에 입력 값 추가
4. P,L,D,B 연산 수행
5. P는 왼쪽에 입력 값이 추가되므로 left 리스트에 값 추가
6. L은 커서가 왼쪽으로 가므로 left 리스트의 마지막 값을 right 리스트의 앞에 값 추가
7. D는 커서가 오른쪽으로 가므로 left 리스트의 마지막 값에 right 리스트 앞 값을 추가
8. B는 왼쪽 문자를 삭제하므로 left 리스트의 마지막 값 삭제 후 right 리스트는 그대로
9. left, right 리스트 출력

```python
string = input()
left, right = [], []
for c in string:
    left.append(c)
n = int(input())
for _ in range(n):
    c = input()
    if c == "P":
        text = input()
        left.append(text)
    elif c == "L":
        text = left.pop()
        right.insert(0,text)
    elif c == "D":
        text = right.pop
        left.append(text)
    elif c == "B":
        left.pop() 
    
for l in left:
    print(l, end="")
    
for r in right:
    print(r, end="")
    
    
        
```

```python
# 인터넷 답안 비교해보기
```

```python
string_list = list(input())
cursor = len(string_list)

for _ in range(int(input())):
    command = list(input().split())
    if command[0] == 'P':
        string_list.insert(cursor, command[1])
        cursor += 1
  
    elif command[0] == 'L':
        if cursor > 0:
            cursor -= 1

    elif command[0] =='D':
        if cursor < len(string_list):
            cursor += 1
  
    else:
        if cursor > 0:
            string_list.remove(string_list[cursor-1])

print(''.join(string_list))
```
